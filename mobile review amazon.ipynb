```python
import requests
from bs4 import BeautifulSoup as bs
```

```python
url='https://www.amazon.in/Apple-iPhone-Pro-Max-256/product-reviews/B0CHWV2WYK/ref=cm_cr_dp_d_show_all_btm?ie=UTF8&reviewerType=all_reviews'
page=requests.get(url)
page
```

```python
page.content
```

```python
soup=bs(page.content,'html.parser')
soup
```

```python
names=soup.find_all('span',class_='a-profile-name')
names
```

```python
len(names)
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(names, columns=['customer Names'])
df
```

```python
df=df.drop_duplicates(ignore_index=True)
df
```

```python
r_title=soup.find_all('a',class_='review-title')
r_title
```

```python
len(r_title)
```

```python
review_title=[]
for i in range(0,len(r_title)):
    review_title.append(r_title[i].get_text())
    review_title
```

```python
print(review_title)
```

```python
dates=soup.find_all('span',class_='review-date')[2:]
dates
```

```python
len(dates)
```

# date filtering

```python
review_dates=[]
for i in range(0,len(dates)):
    review_dates.append(dates[i].get_text().replace('Reviewed in India on ',''))
    review_dates
```

```python
print(review_dates)
```

```python
df['Review Title']=review_title
df['Date']=review_dates
df
```
